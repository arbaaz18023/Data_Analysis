```
# port length


def lnt(lt):
  count = 0;
  for i in lt:
    if(len(i)>1):
      if(i[0]!="#" and i[0]!="@"):
        ln = i.split(".");
        counter = 0;
        for k in ln:
          if(len(k)!=0):
            counter = counter+1
        if(counter<=1):
          count = count+1
  return count



hashf = {}
hashl = {}
for i in parlerlt:
    temp = str(i[7])
   # temp = temp.replace("\n"," ")
    temp = temp.replace("\\n"," ")
    lt = temp.split(" ")
    if(len(lt)>0):
        num =lnt(lt)
        for j in lt:
            if(len(j)>3 and j[0]=="#"):
              j = j.lower()
              if(j in hashf):
                temp3 = hashf[j]+1
                hashf[j] = temp3
              if(j in hashl):
                hashl[j] = hashl[j] + num
              else:
                hashf[j] = 1
                hashl[j] = num
sort_orders = sorted(hashf.items(), key=lambda x: x[1], reverse=True) 
print(sort_orders)
top_10_fr = []
top_10_hash = []
for i in range(10):
  top_10_fr.append(sort_orders[i][1])
  top_10_hash.append(sort_orders[i][0])
plt.bar(top_10_hash, top_10_fr,color = 'g',align = 'center', alpha = 0.5)
plt.ylabel('frequency')
plt.xticks(rotation = 90)
plt.xlabel('hashtags')
plt.show()
#sort_orders1 = sorted(hashl.items(), key=lambda x: x[1], reverse=True) 
aver = {}
for i in hashl:
  temp = hashl[i]
  aver[i] = temp//hashf[i]
sort_orders = sorted(aver.items(), key=lambda x: x[1], reverse=True)
top_10_fr = []
top_10_hash = []
for i in range(10):
  top_10_fr.append(sort_orders[i][1])
  top_10_hash.append(sort_orders[i][0])
plt.bar(top_10_hash, top_10_fr,color = 'g',align = 'center', alpha = 0.5)
plt.ylabel('Postlen')
plt.xticks(rotation = 90)
plt.xlabel('hashtags')
plt.show()
print(sort_orders)
```
